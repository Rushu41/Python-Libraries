Box Plot

```python
box_df = sns.load_dataset('tips')
box_df.head()
```

```python
sns.set(style ='whitegrid')
plt.figure(figsize=(12,6))
sns.boxplot(x ='day', y = 'total_bill', data = box_df,hue = 'sex')
plt.show()
```

```python
sns.set(style ='whitegrid')
plt.figure(figsize=(12,6))
sns.boxplot(x ='day', y = 'total_bill', data = box_df,hue = 'sex',color='g', order =['Sat','Fri','Sun','Thur'])
plt.show()
```

```python
sns.set(style ='whitegrid')
plt.figure(figsize=(12,6))
sns.boxplot(x ='day', y = 'total_bill', data = box_df, color='w', order =['Sat','Fri','Sun','Thur'],
            showmeans = True, linewidth= 3 ,linecolor='r',
            meanprops = {'marker': '+', 'markeredgecolor':'b' , 'markersize': 10})
plt.show()
```
